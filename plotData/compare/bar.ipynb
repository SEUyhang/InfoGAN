```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager

my_font = font_manager.FontProperties(fname="/home/yhang/assets/SimHei.ttf")
plt.rcParams['font.sans-serif'] = ['SimHei']

labels = ['SVM', '1D-CNN','RNN', 'LSTM','CNN + self-attention', 'Ours']
# SVM 1D-CNN RNN LSTM  CNN + self-attention Ours
# Accuracy = np.array([88.53, 87, 78.3, 90.8, 96.5, 98.69])
# Recall = np.array([87.56, 93.5, 83.8, 93.5, 95.1, 97.94])
# Precision = np.array([93.44, ])
# TPR = np.array([81.80, 93.5, 83.8, 93.5, 95.1, 97.94])
# TNR = np.array([93.44, 82.2, 74.2, 25.7, 16.2, 99.48])
# FPR = np.array([6.55, 17.8, 25.7, 11.7, 2.1, 0.51])
# FNR = np.array([18.19, 6.4, 16.2, 6.5, 4.9, 2.05])

# SVM = [88.53, 81.80, 93.44, 6.55, 18.19]
# CNN = [87, 93.5, 82.2, 17.8, 6.4]
# RNN = [78.3, 83.8, 74.2, 25.7, 16.2]
# LSTM = [90.8, 93.5, 88.3, 11.7, 6.5]
# self_attention = [96.5, 95.1, 97.9, 2.1, 4.9]
# Ours = [98.69, 97.94, 99.48, 0.51, 2.05]

data = np.array([[88.53, 81.80, 93.44, 6.55, 18.19],
                 [87, 93.5, 82.2, 17.8, 6.4],
                 [78.3, 83.8, 74.2, 25.7, 16.2],
                 [90.8, 93.5, 88.3, 11.7, 6.5],
                 [96.5, 95.1, 97.9, 2.1, 4.9],
                 [98.69, 97.94, 99.48, 0.51, 2.05]])

num_methods = data.shape[0]
num_metrics = data.shape[1]

fig, ax = plt.subplots()

index = np.arange(num_metrics)
bar_width = 0.1

opacity = 0.75

colors = ['#ffff33', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#e41a1c']
labels = ['SVM', 'CNN', 'RNN', 'LSTM', 'self-attention', 'Ours']

for i in range(num_methods):
    rects = ax.bar(index + i*bar_width, data[i], bar_width,
                   alpha=opacity, color=colors[i], label=labels[i])

ax.set_xlabel('评价指标',fontproperties=my_font)
ax.set_ylabel('检测成功率(单位 ：%)',fontproperties=my_font)
ax.set_title('GNSS欺骗干扰检测对比图',fontproperties=my_font)
ax.set_xticks(index + bar_width*(num_methods-1) / 2)
ax.set_xticklabels(('准确率', '真阳性率', '真阴性率', '假阳率', '假阴率'),fontproperties=my_font)
ax.legend()

fig.tight_layout()
# 保存图形为PNG格式
plt.savefig('GNSS_compare.png', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_groups = 5
n_methods = 6

# 每种方法在每种度量上的平均得分
means = np.array([[20, 35, 30, 35, 27],
                  [25, 32, 34, 20, 25],
                  [22, 30, 32, 25, 28],
                  [28, 33, 31, 22, 26],
                  [24, 28, 36, 27, 29],
                  [26, 29, 33, 23, 30]])

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.1

opacity = 0.8

colors = ['b', 'r', 'g', 'c', 'm', 'y']
labels = ['Method 1', 'Method 2', 'Method 3', 'Method 4', 'Method 5', 'Method 6']

for i in range(n_methods):
    rects = ax.bar(index + i*bar_width, means[i], bar_width,
                   alpha=opacity, color=colors[i], label=labels[i])

ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Scores by method and metric')
ax.set_xticks(index + bar_width*(n_methods-1) / 2)
ax.set_xticklabels(('Metric A', 'Metric B', 'Metric C', 'Metric D', 'Metric E'))
ax.legend()

fig.tight_layout()
plt.show()
```
